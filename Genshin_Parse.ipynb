```python
def format_field(x):
    return '"{}"'.format(x) if isinstance(x, str) else str(x)

def gen_insert_query(df, df_columns, db_columns, table_name):
    if not isinstance(df, pd.DataFrame):
        df = pd.DataFrame(df)
        
    data = ', '.join(df[df_columns].apply(lambda x: '({})'.format(', '.join(map(format_field, x))), axis=1))
    query = "INSERT INTO {} ({}) VALUES {}".format(table_name, ', '.join(db_columns), data)
    return query
```
